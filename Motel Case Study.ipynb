```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
hotel_type=pd.read_csv("Accomodation_and_HotelType.csv")
product_data=pd.read_csv("Product_Data.csv")
```

```python
hotel_type.head()
```

```python
product_data.head()
```

```python
product_data.info()
```

```python
product_data.isna().sum()
```

```python
merged_data=pd.merge(hotel_type,product_data,how='inner',on='hotel_type_id')
```

```python
merged_data
```

```python
merged_data.info()
```

```python
merged_data.isna().sum()
```

```python
merged_data['booking_date']=pd.to_datetime(merged_data['booking_date'])
merged_data['checkin_date']=pd.to_datetime(merged_data['checkin_date'])
merged_data['checkout_date']=pd.to_datetime(merged_data['checkout_date'])
```

```python
merged_data.info()
```

```python
merged_data['ADR_USD'] = merged_data['ADR_USD'].str.strip().str.replace(',', '')
```

```python
merged_data['ADR_USD'] = merged_data['ADR_USD'].astype(float)
```

```python
merged_data.info()
```

```python
merged_data[['ADR_USD','star_rating']].describe()
```

Understanding the Days by Bookings, Checkins and Checkouts


```python
#Understanding the Checkins by weekdays
merged_data['Check_in_Day']=merged_data['checkin_date'].dt.day_name()
merged_data['Check_in_Day']
```

```python
#Understanding which day has more Checkins
day_booking_count=merged_data.groupby('Check_in_Day')['ADR_USD'].count().reset_index(name='Count_by_Day').sort_values(by='Count_by_Day',ascending=False)
day_booking_count
```

```python
plt.figure(figsize=(10, 6))
plt.bar(day_booking_count['Check_in_Day'], day_booking_count['Count_by_Day'], color='skyblue')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Checkins')
plt.title('Number of Checkins by Day of the Week')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Understanding the Checkouts by weekdays
merged_data['Check_out_Day']=merged_data['checkout_date'].dt.day_name()
merged_data['Check_out_Day']
```

```python
#Understanding which day has more Checkouts
day_checkout_count=merged_data.groupby('Check_out_Day')['ADR_USD'].count().reset_index(name='Count_by_Day').sort_values(by='Count_by_Day',ascending=False)
day_checkout_count
```

```python
plt.figure(figsize=(10, 6))
plt.bar(day_checkout_count['Check_out_Day'], day_checkout_count['Count_by_Day'], color='skyblue')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Checkouts')
plt.title('Number of Checkouts by Day of the Week')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Understanding to see which day has more bookings
merged_data['booking_date_Day']=merged_data['booking_date'].dt.day_name()
merged_data['booking_date_Day']
```

```python
#Understanding to see which day has more bookings
day_booking_count=merged_data.groupby('booking_date_Day')['ADR_USD'].count().reset_index(name='Count_by_Day').sort_values(by='Count_by_Day',ascending=False)
day_booking_count
```

```python
plt.figure(figsize=(10, 6))
plt.bar(day_booking_count['booking_date_Day'], day_booking_count['Count_by_Day'], color='skyblue')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Bookings')
plt.title('Number of Bookings by Day of the Week')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```

Understanding Price charged by the no of days

```python
#Understanding why there are more bookings on Friday by looking at the average prices by day
day_booking_count=merged_data.groupby('booking_date_Day')['ADR_USD'].mean().reset_index(name='Average_by_Day').sort_values(by='Average_by_Day',ascending=False)
day_booking_count
```

```python
plt.figure(figsize=(10, 6))
plt.bar(day_booking_count['booking_date_Day'], day_booking_count['Average_by_Day'], color='skyblue')
plt.xlabel('Day of the Week')
plt.ylabel('Average booking by AUD')
plt.title('Average booking by AUD by Day of the Week')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#How does booking date relate to check-in date and ADR (e.g., bookings made closer to check-in dates)?
merged_data['Days_diff_booking_checkin']=merged_data['checkin_date']-merged_data['booking_date']
merged_data
```

```python
#Understanding the AUD by days before booked
Price_by_no_of_days=merged_data.groupby('Days_diff_booking_checkin')['ADR_USD'].mean().reset_index(name='Avg_price').sort_values(by='Avg_price',ascending=False)
Price_by_no_of_days['Days_diff_booking_checkin'] = Price_by_no_of_days['Days_diff_booking_checkin'].apply(lambda x: x.days)
Price_by_no_of_days
```

```python
plt.figure(figsize=(10, 6))
plt.bar(Price_by_no_of_days['Days_diff_booking_checkin'], Price_by_no_of_days['Avg_price'], color='skyblue')
plt.xlabel('No of Days')
plt.ylabel('Average price by AUD')
plt.title('Average price by no of days')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```

Understanding the days, chain, rating by price booking date 

```python
chain_hotels['booking_date_Day_no']=chain_hotels['booking_date'].dt.dayofweek
chain_hotels
```

```python
chain_day_rating_hotels=chain_hotels.groupby(['hotel_type','booking_date_Day_no','booking_date_Day','star_rating'])['ADR_USD'].mean().reset_index(name='Average_price').sort_values(by=['star_rating','booking_date_Day_no'],ascending=[False,True])
chain_day_rating_hotels
```

```python
plt.figure(figsize=(14, 8))


for star_rating in sorted(chain_day_rating_hotels['star_rating'].unique()):
    subset = chain_day_rating_hotels[chain_day_rating_hotels['star_rating'] == star_rating]
    plt.plot(subset['booking_date_Day_no'], subset['Average_price'], marker='o', label=f'Star Rating {star_rating}')


plt.xlabel('Day of the Week (Monday=0, Tuesday=1, ..., Sunday=6)', fontsize=14)
plt.ylabel('Average Price (ADR USD)', fontsize=14)
plt.title('Average Price by Booking Day and Star Rating and Chain', fontsize=16)
plt.xticks(range(7), ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.legend(title='Star Rating', title_fontsize='13', fontsize='12')
plt.grid(True)


plt.tight_layout()
plt.show()
```

```python
non_chain_hotels['booking_date_Day_no']=non_chain_hotels['booking_date'].dt.dayofweek
non_chain_hotels
```

```python
plt.figure(figsize=(14, 8))


for star_rating in sorted(non_chain_day_rating_hotels['star_rating'].unique()):
    subset = non_chain_day_rating_hotels[non_chain_day_rating_hotels['star_rating'] == star_rating]
    plt.plot(subset['booking_date_Day_no'], subset['Average_price'], marker='o', label=f'Star Rating {star_rating}')


plt.xlabel('Day of the Week (Monday=0, Tuesday=1, ..., Sunday=6)', fontsize=14)
plt.ylabel('Average Price (ADR USD)', fontsize=14)
plt.title('Average Price by Booking Day and Star Rating and Non Chain', fontsize=16)
plt.xticks(range(7), ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.legend(title='Star Rating', title_fontsize='13', fontsize='12')
plt.grid(True)


plt.tight_layout()
plt.show()

```
